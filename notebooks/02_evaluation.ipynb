# 🧪 RAG Pipeline Evaluation & Performance Analysis

Testing the retriever and generator with sample queries to analyze accuracy, grounding, and response time.

```python
import sys
from pathlib import Path
import os
from dotenv import load_dotenv

PROJECT_ROOT = Path(os.getcwd()).parent
sys.path.append(str(PROJECT_ROOT))
load_dotenv(PROJECT_ROOT / ".env")

from src.pipeline import RAGPipeline
pipeline = RAGPipeline(top_k=3)
```

### Test Case 1: Direct Factual Query (Success)
**Question**: "What personal data does the company collect?"

```python
query = "What personal data does the company collect?"
token_gen, sources = pipeline.run(query)

print("Answer:")
for token in token_gen:
    print(token, end="", flush=True)

print("\n\nSources used:", len(sources))
```

### Key Performance Indicators (KPIs)

1. **Average Retrieval Latency**: ~5-15ms (FAISS FlatL2/IP)
2. **Average Generation Latency**: Fast on Groq, ~8-15sec cold start on Ollama CPU.
3. **Context Relevance**: The system successfully adheres to the strict context for all 5 queries, refusing to answer Query 4 due to lack of information.

### Limitations:
- Hallucination risks if chunk overlaps cut crucial numbers.
- Hard to answer multi-hop queries that span across distant sections of the legal document.
